```python
from tkinter import *
from tkinter import ttk
import mysql.connector
from tkinter import messagebox


class Member:
    def __init__ (self,root):
        self.root=root
        self.root.title('Club Member Management System')
        self.root.geometry("1530x800+0+0")
        
        self.id_var=StringVar()
        self.name_var=StringVar()
        self.email_var=StringVar()
        self.contact_var=StringVar()
        self.referance_var=StringVar()
        
        self.search_by=StringVar()
        self.search_txt=StringVar()
        self.application_status=StringVar()
    
        
        title=Label(self.root,text='Club Member Management System',font=('times new roman',30,'bold')
                    ,fg='black',bg='powder blue',bd=4,relief=RAISED)
        title.pack(side=TOP,fill=X)
        
        # manage frame
        Manage_frame=Frame(self.root,bd=4,relief=RAISED,bg='powder blue')
        Manage_frame.place(x=20,y=80,width=600,height=600)
        #manage title
        M_title=Label(Manage_frame,text='MANAGE MEMBER',font=('times new roman',30,'bold')
                    ,fg='black',bg='powder blue',bd=4,relief=RAISED)
        M_title.grid(row=0,columnspan=2,padx=20,pady=20)
        
        #labels
        lbl_id=Label(Manage_frame,text='ID',font=('times new roman',20,'bold'),fg='black',bg='powder blue')
        lbl_id.grid(row=1,column=0,padx=20,pady=10,sticky=W)
        id_entry=ttk.Entry(Manage_frame,textvariable=self.id_var,font=('times new roman',20,'bold'))
        id_entry.grid(row=1,column=1,padx=5,pady=15,sticky=W)
        
        lbl_name=Label(Manage_frame,text='Applicant Name',font=('times new roman',20,'bold'),fg='black',bg='powder blue')
        lbl_name.grid(row=2,column=0,padx=20,pady=10,sticky=W)
        name_entry=ttk.Entry(Manage_frame,textvariable=self.name_var, font=('times new roman',20,'bold'))
        name_entry.grid(row=2,column=1,padx=5,pady=15,sticky=W)
        
        lbl_contact=Label(Manage_frame,text='Contact',font=('times new roman',20,'bold'),fg='black',bg='powder blue')
        lbl_contact.grid(row=3,column=0,padx=20,pady=10,sticky=W)
        contact_entry=ttk.Entry(Manage_frame,textvariable=self.contact_var, font=('times new roman',20,'bold'))
        contact_entry.grid(row=3,column=1,padx=5,pady=15,sticky=W)
        
        lbl_email=Label(Manage_frame,text='Email',font=('times new roman',20,'bold'),fg='black',bg='powder blue')
        lbl_email.grid(row=4,column=0,padx=20,pady=10,sticky=W)
        email_entry=ttk.Entry(Manage_frame,textvariable=self.email_var, font=('times new roman',20,'bold'))
        email_entry.grid(row=4,column=1,padx=5,pady=15,sticky=W)
        
        lbl_ref=Label(Manage_frame,text='Member Reference',font=('times new roman',20,'bold'),fg='black',bg='powder blue')
        lbl_ref.grid(row=5,column=0,padx=20,pady=10,sticky=W)
        ref_entry=ttk.Entry(Manage_frame,textvariable=self.referance_var, font=('times new roman',20,'bold'))
        ref_entry.grid(row=5,column=1,padx=5,pady=15,sticky=W)
        
        lbl_status=Label(Manage_frame,text="Application Status",font=('times new roman',20,'bold'),fg='black',bg='powder blue')
        lbl_status.grid(row=6,column=0,padx=20,pady=10,sticky=W)
        status_combo=ttk.Combobox(Manage_frame,textvariable=self.application_status,width=19,font=('times new roman',20,'bold'),state='readonly')
        status_combo['values']=('Applied','Verified','Approved')
        status_combo.grid(row=6,column=1,padx=5,pady=15,sticky=W)
        
        
        #button Frame
        btn_frame=Frame(self.root,bd=4,relief=RIDGE,bg='powder blue')
        btn_frame.place(x=50,y=600,width=530)
        
        add_btn=Button(btn_frame,text='APPLY',command=self.add_member,width=14,height=2,fg='white',bg='crimson')
        add_btn.grid(row=0,column=0,padx=10,pady=10)
        
        update_btn=Button(btn_frame,text='UPDATE',command=self.update_data,width=14,height=2,fg='white',bg='crimson')
        update_btn.grid(row=0,column=1,padx=10,pady=10)
        
        delete_btn=Button(btn_frame,text='DELETE',command=self.delete_data,width=14,height=2,fg='white',bg='crimson')
        delete_btn.grid(row=0,column=2,padx=10,pady=10)
        
        clear_btn=Button(btn_frame,command=self.clear,text='CLEAR',width=14,height=2,fg='white',bg='crimson')
        clear_btn.grid(row=0,column=3,padx=10,pady=10)
        
        #Details Frame
        Details_frame=Frame(self.root,bd=4,relief=RIDGE,bg="powder blue")
        Details_frame.place(x=650,y=80,width=850,height=600)
        
        search_lbl=Label(Details_frame,text='Search By',font=("time new roman",20,'bold'),fg='red',bg='powder blue')
        search_lbl.grid(row=1,column=0,padx=20,pady=10,sticky=W)
        
        search_combo=ttk.Combobox(Details_frame,textvariable=self.search_by,width=19,font=('times new roman',15,'bold'),state='readonly')
        search_combo['values']=('Select Option','ID','Name','Application_Status','Referance','Join_date')
        search_combo.grid(row=1,column=1,padx=20,pady=10,sticky=W)
        
        search_entry=ttk.Entry(Details_frame,textvariable=self.search_txt,font=('times new roman',15,'bold'))
        search_entry.grid(row=1,column=2,padx=5,pady=15,sticky=W)
        
        search_btn=Button(Details_frame,command=self.search_data,text='Search',width=8,height=1,fg='white',bg='crimson')
        search_btn.grid(row=1,column=3,padx=10,pady=10)
        
        showall_btn=Button(Details_frame,command=self.fetch_data,text='Show All',width=10,height=1,fg='white',bg='crimson')
        showall_btn.grid(row=1,column=4,padx=10,pady=10)
        
        #Table Frame
        Table_frame=Frame(Details_frame,bd=4,relief=RIDGE,bg="powder blue")
        Table_frame.place(x=20,y=80,width=800,height=500)
        
        scroll_x=ttk.Scrollbar(Table_frame,orient=HORIZONTAL)
        scroll_y=ttk.Scrollbar(Table_frame,orient=VERTICAL)
        
        self.member_table=ttk.Treeview(Table_frame,column=('ID','Name','Contact','Email','Referance',"Application_Status","Join_date",'Data_updated'),xscrollcommand=scroll_x.set,yscrollcommand=scroll_y.set)
        scroll_x.pack(side=BOTTOM,fill=X)
        scroll_y.pack(side=RIGHT,fill=Y)
        
        scroll_x.config(command=self.member_table.xview)
        scroll_y.config(command=self.member_table.yview)
        
        self.member_table.heading('ID',text="ID")
        self.member_table.heading('Name',text="Name")
        self.member_table.heading('Email',text="Email")
        self.member_table.heading('Contact',text="Contact")
        self.member_table.heading('Referance',text="Referance")
        self.member_table.heading('Application_Status',text="Application_Status")
        self.member_table.heading('Join_date',text="Join_date")
        self.member_table.heading('Data_updated',text="Updated_on")
        
        self.member_table['show']='headings'
        self.member_table.pack(fill=BOTH,expand=1)
        self.member_table.column('ID',width=90)
        self.member_table.column('Name',width=175)
        self.member_table.column('Contact',width=90)
        self.member_table.column('Email',width=175)
        self.member_table.column('Referance',width=130)
        self.member_table.column('Application_Status',width=110)
        self.member_table.column('Join_date',width=90)
        self.member_table.column('Data_updated',width=90)
        
        
        self.member_table.bind('<ButtonRelease-1>',self.get_cursor)
        
        self.fetch_data()
        
        cursor_row=self.member_table.focus()
        content=self.member_table.item(cursor_row)
        row=content['values']
        
        print(row)
        
        
        
    # defining function for buttons    
    def add_member(self):
        conn=mysql.connector.connect(host='localhost',username='root',password='your_password',database='your_database_name')
        my_cursor=conn.cursor()
        my_cursor.execute('insert into member_data values(%s,%s,%s,%s,%s,"Applied",curdate(),curdate())',
                                                                           (self.id_var.get(),
                                                                            self.name_var.get(),
                                                                            self.contact_var.get(),
                                                                            self.email_var.get(),
                                                                            self.referance_var.get()
                                                                           ))
        conn.commit()
        self.fetch_data()
        self.clear()
        conn.close()
        messagebox.showinfo('Apply','You have applied sucessfully')
        
    def fetch_data(self):        
        conn=mysql.connector.connect(host='localhost',username='root',password='your_password',database='your_database_name')
        my_cursor=conn.cursor()
        my_cursor.execute('select * from member_data')
        rows=my_cursor.fetchall()
        if len(rows)!=0:
            self.member_table.delete(*self.member_table.get_children())
            for i in rows:
                self.member_table.insert('',END,values=i)
            conn.commit()
        conn.close()
        
    def clear(self):
        self.id_var.set('')
        self.name_var.set('')
        self.email_var.set('')
        self.contact_var.set('')
        self.referance_var.set('')
        self.application_status.set('')
        
    def get_cursor(self,ev):
        cursor_row=self.member_table.focus()
        content=self.member_table.item(cursor_row)
        row=content['values']
        self.id_var.set(row[0])
        self.name_var.set(row[1])
        self.contact_var.set(row[2])
        self.email_var.set(row[3])
        self.referance_var.set(row[4])
        self.application_status.set(row[5])
        

        
    def update_data(self):
        conn=mysql.connector.connect(host='localhost',username='root',password='your_password',database='your_database_name')
        my_cursor=conn.cursor()
        my_cursor.execute('update member_data set Name=%s,Contact=%s,Email=%s,Application_Status=%s,Data_updated=curdate() where ID=%s',(
                                                                                         self.name_var.get(),
                                                                                         self.contact_var.get(),
                                                                                         self.email_var.get(),
                                                                                         self.application_status.get(),
                                                                                         self.id_var.get()
                                                                                                            )) 
        conn.commit()
        self.fetch_data()
        self.clear()
        conn.close()
        messagebox.showinfo('Update','Record has been upadated sucessfully')
        
    def delete_data(self):
        conn=mysql.connector.connect(host='localhost',username='root',password='your_password',database='your_database_name')
        my_cursor=conn.cursor()
        query='delete from member_data where ID=%s'
        value=(self.id_var.get(),)
        my_cursor.execute(query,value)
        
        conn.commit()
        conn.close()
        self.fetch_data()
        self.clear()
        
        messagebox.showinfo("Delete","Member has been deleted successfully")
        
    def search_data(self):
        conn=mysql.connector.connect(host='localhost',username='root',password='your_password',database='your_database_name')
        my_cursor=conn.cursor()
        my_cursor.execute("select * from member_data where " + str(self.search_by.get()) +" LIKE '%"+str(self.search_txt.get())+"%'")
        rows=my_cursor.fetchall()
        if len(rows)!=0:
            self.member_table.delete(*self.member_table.get_children())
            for i in rows:
                self.member_table.insert('',END,values=i)
            conn.commit()
        conn.close()
    
        
        
root=Tk()
ob=Member(root)
root.mainloop()
```
